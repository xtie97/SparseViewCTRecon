```python
# import libraries
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [7, 7]
import numpy as np
from utils import read_binary_dataset
import torch
import math
from torch_radon import Radon, RadonFanbeam

# define some constants
batch_size = 16
img_size = 512
n_angles = 984
device = torch.device('cuda')

# load phantom and move it to GPU
img = read_binary_dataset('./intermediate/target_0000.raw', dim=(img_size, img_size))
batch_size = img.shape[0]
x = torch.FloatTensor(img).to(device)

_ = plt.imshow(img[0,:], cmap="gray") 
```

```python
# define measuring angles
angles = np.linspace(0, np.pi, n_angles, endpoint=False)
det_count = 888 
#det_count = int(np.sqrt(2)*img_size + 0.5)
```

```python
# instantiate Radon transform
radon = Radon(img_size, angles, clip_to_circle=False, det_count=det_count)

sinogram = radon.forward(x)
bp = radon.backprojection(sinogram)
filtered_sinogram = radon.filter_sinogram(sinogram)
index = 1
fbp = radon.backprojection(filtered_sinogram)

plt.figure()
_ = plt.imshow(sinogram[index,:].cpu(), cmap="gray") 
plt.figure()
_ = plt.imshow(bp[index,:].cpu(), cmap="gray") 
plt.figure()
_ = plt.imshow(fbp[index,:].cpu(), cmap="gray") 
plt.figure()
_ = plt.imshow((fbp[index,:] - x[index,:]).cpu(), cmap="gray") 
```

```python
# instantiate Radon Fan-beam transform
n_angles = 984 # number of view angles 
det_count = 888 # number of detector columns
nfan = 55.0059 / 180 * math.pi # degree of fan angle -> rad 
image_size = 512 
SAD = 538.52 # source to axis distance (mm) 
SDD = 946.746 # source to detector distance (mm)
det_spacing = SDD * math.tan(nfan/2) * 2 / det_count

fanbeam_angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)
radon_fanbeam = RadonFanbeam(image_size, 
                             fanbeam_angles, 
                             source_distance=SAD, 
                             det_distance=SDD-SAD, 
                             det_count=det_count, 
                             det_spacing=det_spacing)
# clip_to_circle=False


sinogram = radon_fanbeam.forward(x)
bp = radon_fanbeam.backprojection(sinogram)
fbp = radon.backprojection(filtered_sinogram)

index = 1

plt.figure()
_ = plt.imshow(sinogram[index,:].cpu(), cmap="gray") 
plt.figure()
_ = plt.imshow(bp[index,:].cpu(), cmap="gray") 
plt.figure()
_ = plt.imshow(fbp[index,:].cpu(), vmin=-200+1024, vmax=200+1024, cmap="gray") 
plt.figure()
_ = plt.imshow((fbp[index,:] - x[index,:]).cpu(), vmin=-50, vmax=+50, cmap="gray") 
```
